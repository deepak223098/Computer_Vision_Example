# Play with Images

### 1. Smoothing and Blurring

### 1.1 Homogenious Filter
###### kernal is K = 1/n * matrix
###### matrix = height*length(5,5)
###### 1/25*[5X5]{matrix is filled with all ones}
##### Low pass filter helps in removing noises in the blurring the iamges
##### High pass filter helps in fidning edges in the images

```python
%matplotlib inline
#%matplotlib tk 
#%matplotlib
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("opencv-logo.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernal = np.ones((5,5),np.float32)/25
dest = cv2.filter2D(img,-1,kernal)
titles = ["img","2D Convolution"]
images = [img,dest]
for i in range(len(titles)):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### 1.2 Blue Images using cv2.blur

```python
%matplotlib qt
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("opencv-logo.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernal = np.ones((5,5),np.float32)/25
dest = cv2.filter2D(img,-1,kernal)
blur = cv2.blur(img,(3,3))
titles = ["img","2D Convolution","blur"]
images = [img,dest,blur]
for i in range(len(titles)):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### 1.3 Blur Images using gaussian blur
#### Gaussian filter is nothing but using the different weight kernal in both x and y direction. Gaussian blue is use to remove high frequency noise in the images
#### pixel located the in side are giving less weight and pixel located in the center are giving more weight in a given image matrix

```python
%matplotlib qt
import cv2
import numpy as np
from matplotlib import pyplot as plt
#img = cv2.imread("opencv-logo.png")
img = cv2.imread("Halftone_Gaussian_Blur.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernal = np.ones((5,5),np.float32)/25
dest = cv2.filter2D(img,-1,kernal)
blur = cv2.blur(img,(3,3))
gblur = cv2.GaussianBlur(img,(5,5),0)
titles = ["img","2D Convolution","blur","gblur"]
images = [img,dest,blur,gblur]
for i in range(len(titles)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### 1.4 Blur Images using Median filter. 
Median filter is something that replace each pixel's value with the median of its neighbouring pixels, This method is great when dealing with "salt and pepper noise"

```python
%matplotlib qt
import cv2
import numpy as np
from matplotlib import pyplot as plt
#img = cv2.imread("opencv-logo.png")
#img = cv2.imread("Halftone_Gaussian_Blur.jpg")
img = cv2.imread("water.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernal = np.ones((5,5),np.float32)/25
dest = cv2.filter2D(img,-1,kernal)
blur = cv2.blur(img,(3,3))
gblur = cv2.GaussianBlur(img,(5,5),0)
median = cv2.medianBlur(img,5) # here 5 is the kernal size and it should be odd
titles = ["img","2D Convolution","blur","gblur","median"]
images = [img,dest,blur,gblur,median]
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### 1.5 Blur Images using Bilateral filter. 
#### images in which we need to preserve the edges sharp with smoothing then use this Bilateral filter

```python
%matplotlib qt
import cv2
import numpy as np
from matplotlib import pyplot as plt
#img = cv2.imread("opencv-logo.png")
#img = cv2.imread("Halftone_Gaussian_Blur.jpg")
img = cv2.imread("lena.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernal = np.ones((5,5),np.float32)/25
dest = cv2.filter2D(img,-1,kernal)
blur = cv2.blur(img,(3,3))
gblur = cv2.GaussianBlur(img,(5,5),0)
median = cv2.medianBlur(img,5) # here 5 is the kernal size and it should be odd
bilateralFilter = cv2.bilateralFilter(img,9,75,75)          
titles = ["img","2D Convolution","blur","gblur","median","bilateralFilter"]
images = [img,dest,blur,gblur,median,bilateralFilter]
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

# 2. Image Gradients and Edge Detection

An image gradient is a directional change in the intensity or color in an image.

### 2.1 Laplassian Gradeint

```python
%matplotlib
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("messi5.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
lap = cv2.Laplacian(img,cv2.CV_64F,ksize = 3)
lap =  np.uint8(np.absolute(lap))
titles = ["img","Laplacian"]
images = [img,lap]
for i in range(len(titles)):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### 2.2 Sobal X Gradient

```python
%matplotlib
import cv2
import numpy as np
from matplotlib import pyplot as plt
#img = cv2.imread("messi5.jpg")
img = cv2.imread("sudoku.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

lap = cv2.Laplacian(img,cv2.CV_64F,ksize = 3)
lap =  np.uint8(np.absolute(lap))

soblx = cv2.Sobel(img,cv2.CV_64F,1,0)#1order of derative of x
sobly = cv2.Sobel(img,cv2.CV_64F,0,1)#1 order of derative of y

sobelx =  np.uint8(np.absolute(soblx))
sobely =  np.uint8(np.absolute(sobly))

titles = ["img","Laplacian","sobelx","sobely"]
images = [img,lap,sobelx,sobely]
for i in range(len(titles)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
#img = cv2.imread("messi5.jpg")
img = cv2.imread("sudoku.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

lap = cv2.Laplacian(img,cv2.CV_64F,ksize = 3)
lap =  np.uint8(np.absolute(lap))

soblx = cv2.Sobel(img,cv2.CV_64F,1,0)#1order of derative of x
sobly = cv2.Sobel(img,cv2.CV_64F,0,1)#1 order of derative of y

sobelx =  np.uint8(np.absolute(soblx))
sobely =  np.uint8(np.absolute(sobly))
sobelcombined = cv2.bitwise_or(sobelx,sobely)
titles = ["img","Laplacian","sobelx","sobely","sobelcombined"]
images = [img,lap,sobelx,sobely,sobelcombined]
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### 2.3 Edge detection using canny method

The Canny edge detector is an edge detection operator  that uses a multi-stage alogrithm to detect a wide range pf edges in images.
The Canny edge detection algorithm is composed of 5 steps:
1. Noise reduction
2. Gradient calculation
3. Non-maxmimum supression
4. Double threshold
5. Edge Tracking by hysteresis

```python
%matplotlib inline
#%matplotlib tk 
#%matplotlib
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("lena.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(img,100,200)
titles = ["img","canny"]
images = [img,canny]
for i in range(len(titles)):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

### 2.3.1 Edge detection using canny method using Track bar

```python
#%matplotlib inline
#%matplotlib qt 
import cv2
import numpy as np
def nothing(x):
    pass
cv2.namedWindow("Tracker")
#cv2.resizeWindow('Tracker', 500, 500)
cv2.createTrackbar("th1","Tracker",0,500,nothing)
cv2.createTrackbar("th2","Tracker",0,500,nothing)
img = cv2.imread("messi5.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
while True:
    th_1 = cv2.getTrackbarPos("th1","Tracker")
    th_2 = cv2.getTrackbarPos("th2","Tracker")
    canny = cv2.Canny(img,th_1,th_2)
    cv2.imshow("img",img)
    cv2.imshow("canny",canny)
    key = cv2.waitKey(1)
    if key ==27:
        break
cv2.destroyAllWindows()
```
