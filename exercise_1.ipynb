#  draw a line ,circle and rectangle on a image

```python
# image = cv2.imread(impath)
image = np.zeros([512,512,3],np.uint8)
print("image.shape {0}".format(image.shape))
cv2.line(image,(0,0),(255,255),(255,0,0),2)
cv2.line(image,(0,255),(255,255),(255,0,0),2)
cv2.line(image,(0,0),(0,255),(255,0,0),2)
cv2.arrowedLine(image,(256,256),(511,511),(0,255,0),2)
cv2.line(image,(50,50),(305,305),(0,0,255),2)
cv2.line(image,(50,305),(305,305),(0,0,255),2)
cv2.line(image,(50,50),(50,305),(0,0,255),2)

# rectangle
cv2.rectangle(image,(384,0),(510,128),(0,255,0),5)

# circle
cv2.circle(image,(447,63),63,(255,0,0),-1)

font = cv2.FONT_HERSHEY_COMPLEX_SMALL
# put text
image = cv2.putText(image,"ShinChan",(100,500),font,2,(255,255,255),2,cv2.LINE_4)

# show image
cv2.imshow("frame", image)

# kill image
kill(image)
```

```python
impath = r"D:\Study\Text_OCR\output\single_pdfs\width_view.jpg"
import cv2
import numpy as np
%matplotlib
#%matplotlib inline
from matplotlib import pyplot as plt
img = cv2.imread(impath,cv2.IMREAD_GRAYSCALE)
_,mask = cv2.threshold(img, 220,255,cv2.THRESH_BINARY_INV)
kernal = np.ones((2,2),np.uint8)
dilation = cv2.dilate(mask,kernal,iterations = 3)
erosion = cv2.erode(mask,kernal,iterations=1)
titles = ['images',"mask","dilation","erosion"]
images = [img,mask,dilation,erosion]
for i in range(len(titles)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
plt.show()
```

```python
%matplotlib 
import cv2
import numpy as np
from matplotlib import pyplot as plt
impath = r"D:\Study\Text_OCR\output\single_pdfs\width_view.jpg"
#img = cv2.imread("messi5.jpg")
img = cv2.imread(impath)
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

lap = cv2.Laplacian(img,cv2.CV_64F,ksize = 3)
lap =  np.uint8(np.absolute(lap))

soblx = cv2.Sobel(img,cv2.CV_64F,1,0)#1order of derative of x
sobly = cv2.Sobel(img,cv2.CV_64F,0,1)#1 order of derative of y

sobelx =  np.uint8(np.absolute(soblx))
sobely =  np.uint8(np.absolute(sobly))
sobelcombined = cv2.bitwise_or(sobelx,sobely)
titles = ["img","Laplacian","sobelx","sobely","sobelcombined"]
images = [img,lap,sobelx,sobely,sobelcombined]
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
